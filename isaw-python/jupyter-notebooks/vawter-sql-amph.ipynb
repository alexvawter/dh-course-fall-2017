```python
#amphs built after 0 AD, ordered by chronogrp start date, o
data = pd.read_sql_query('''select ramphs.title, chrono.start
    from ramphs, chrono
    where ramphs.chronogrp = chrono.chronogrp and chrono.start > 0
    order by chrono.start
    ''',sql_conn)
```

```python
#number of amphitheaters built in each chronogroup
data = pd.read_sql_query ('''select ramphs.chronogrp, count(chronogrp) as "number_of_amphs"
  from  ramphs
  group by chronogrp
  order by count(chronogrp) desc''', sql_conn)
```

```python
#Find all amphitheaters in Italy 
#from sites that Pleiades thinks were founded before 323 BC. 
#Order by earliest date of the amphitheater."
check = pd.read_sql_query('''SELECT ramphs.title, ramphs.modCountry, pleiades.minDate, chrono.start
          FROM ramphs, pleiades, chrono
          WHERE ramphs.pleiades = pleiades.path 
          and ramphs.modCountry = 'Italy'
          and pleiades.minDate < -324 
          and ramphs.chronogrp = chrono.chronogrp
          ORDER BY pleiades.minDate;''', sql_conn)
```

```python
#trying to count the number of Roman amphitheathers built 0 AD or after in Modern countries 
#for which the Pleiades location is precise 
#but I'm only counting the the number of amphs in each modern country regardless of date
df = pd.read_sql_query ('''select ramphs.modCountry as "modern_country", count(modCountry) as "mod_number_amphs",
  chrono.start, pleiades.locationPrecision
  from  ramphs, chrono, pleiades
  where ramphs.pleiades = pleiades.path 
  and ramphs.chronogrp = chrono.chronogrp
  and pleiades.locationPrecision = 'precise'
  and chrono.chronogrp > -1
  group by modCountry
  order by count(modCountry) desc''', sql_conn)
df
```

```python
#number of Roman Amphitheaters in each Modern Country on a choropleth map 
#but I want to make it match the query in the cell above if I figure out how to make it work
data = [dict(
        type = 'choropleth',
        locations = df.modern_country,
        locationmode = 'country names',
        z = df.mod_number_amphs,
        text = df.modern_country,
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
     colorbar = dict(title = 'Roman Amphitheaters'),
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )) )  ]  
fig = dict(data=data)
py.iplot( fig, validate=False,)
```
